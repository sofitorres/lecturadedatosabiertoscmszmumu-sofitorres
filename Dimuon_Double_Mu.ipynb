```python
#panda nos permite manipular el archivo csv https://pandas.pydata.org
import pandas as pd
#numpy nos permite hacer operaciones usando vectores https://numpy.org
import numpy as np
#plotly nos permite crear graficas interactivas https://plotly.com/python/
import plotly.express as px
%matplotlib inline
#usamos la liga que está directamente en la página. 
particles = pd.read_csv('https://cernbox.cern.ch/index.php/s/Ak6DnRo9QhV315Y/download',delimiter=',')
particles.head()
```

```python
# De acuerdo al espectro de masa invariante teórico, podríamos hallar 
# mesones J/psi en un rango de 2.6 < m_i < 3.5 [GeV/c^2]
# fotones en el rango 8 < m_i < 14 [GeV/c^2]
# bosones Z en el rango 1.5 < m_i < 15 [GeV/c^2]
# mesones phi en el rango 0.5 < m_i < 1.5 [GeV/c^2]

# El espectro de masa invariante global es el siguiente

particlesdimuon = particles.query('Q1 != Q2')
fig_1=px.histogram(particlesdimuon['M'], width=700, height=700,
                    title='Espectro de Masa Invariante Global')
fig_1.show()

```

```python
# Cada partícula puede hallarse mejor en ciertas regiones de las variables cinemáticas, analizamos la pseudorapidez eta
# Rangos de: https://iopscience.iop.org/article/10.1088/1748-0221/7/10/P10002
# Buscamos el canal J/psi -> mu+ mu-

# Primero, veamos cómo luce el espectro en esa región de masa

jpsi = particlesdimuon.query('2.6< M <3.5')
fig_jpsi=px.histogram(jpsi['M'], width=600, height=600,
                       title='Espectro de Masa en la región esperada para J/psi')
fig_jpsi.show()

#región 1, si |eta| < 1.3, p_T > 3.3 GeV/c^2
jpsi1 = particlesdimuon.query(
    'pt1 > 3.3 and pt2 > 3.3 and abs(eta1) < 1.3 and abs(eta2) < 1.3 and 2.6 < M < 3.5')
fig_2=px.histogram(jpsi1['M'], width=600, height=600)


fig_2.show()
```

```python
# Buscamos el canal fotón -> mu+ mu-
pho = particlesdimuon.query('8< M <14')
fig_pho=px.histogram(pho['M'], width=550, height=550,
                       title='Espectro de Masa en la región esperada para fotones')
fig_pho.show()

# región 1, si |eta| < 1.6, p_T > 3.5 GeV/c^2

pho1 = particlesdimuon.query(
    'pt1 > 3.5 and pt2 > 3.5 and abs(eta1) < 1.6 and abs(eta2) < 1.6 and 8 < M < 14')
fig_pho1=px.histogram(pho1['M'], width=550, height=550)

# región 2, si 1.6 |eta| < 2.4, p_T > 2.5 GeV/c^2
pho2 = particlesdimuon.query(
    'pt1 > 2.5 and pt2 > 2.5 and 1.6 < abs(eta1) < 2.4 and 1.6 < abs(eta2) < 2.4 and 8 < M < 14')
fig_pho2=px.histogram(pho2['M'], width=550, height=550)

fig_pho1.show()
fig_pho2.show()
```

```python
# Buscamos el canal Z -> mu+ mu-
z = particlesdimuon.query('1.5 < M <15')
fig_z=px.histogram(z['M'], width=550, height=550,
                       title='Espectro de Masa en la región esperada para bosones Z')
fig_z.show()

# región 1, si |eta| < 2.1, p_T > 20 GeV/c^2
z1 = particlesdimuon.query(
    'pt1 > 20 and pt2 > 20 and abs(eta1) < 2.1 and abs(eta2) < 2.1 and 1.5 < M < 15')
fig_z1=px.histogram(z1['M'], width=550, height=550)
fig_z1.show()
```
